```python
# Not final

# Returns data on bed and ventilator usage/availability from imhe_summary.csv as a data frame
def GetEquipData(fipsNum, fileName = 'our_data/imhe_summary.csv'):
    imhe_local_data = pd.read_csv(fileName)
    
    state = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['state']
    
    peak_bed_day_mean = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_bed_day_mean']
    peak_bed_day_lower = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_bed_day_lower']
    peak_bed_day_upper = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_bed_day_upper']
    
    peak_icu_bed_day_mean = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_icu_bed_day_mean']
    peak_icu_bed_day_lower = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_icu_bed_day_lower']
    peak_icu_bed_day_upper = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_icu_bed_day_upper']
    
    peak_vent_day_mean = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_vent_day_mean']
    peak_vent_day_lower = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_vent_day_lower']
    peak_vent_day_upper = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['peak_vent_day_upper']
    
    all_bed_capacity = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['all_bed_capacity']
    icu_bed_capacity = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['icu_bed_capacity']
    all_bed_usage = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['all_bed_usage']
    icu_bed_usage = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['icu_bed_usage']
    
    #print(peak_bed_day_mean, peak_bed_day_lower, peak_bed_day_upper)
    #print(peak_icu_bed_day_mean, peak_icu_bed_day_lower, peak_icu_bed_day_upper)
    #print(peak_vent_day_mean, peak_vent_day_lower, peak_vent_day_upper)
    #print(all_bed_capacity, icu_bed_capacity, all_bed_usage, icu_bed_usage)
    
    outDF = pd.DataFrame({'FIPS':fipsNum, 'state':state,
                        'peak_bed_day_mean':peak_bed_day_mean, 'peak_bed_day_lower':peak_bed_day_lower,
                        'peak_bed_day_upper':peak_bed_day_upper, 'peak_icu_bed_day_mean':peak_icu_bed_day_mean,
                        'peak_icu_bed_day_lower':peak_icu_bed_day_lower, 'peak_icu_bed_day_upper':peak_icu_bed_day_upper,
                        'peak_vent_day_mean':peak_vent_day_mean, 'peak_vent_day_lower':peak_vent_day_lower,
                        'peak_vent_day_upper':peak_vent_day_upper, 'all_bed_capacity':all_bed_capacity,
                        'icu_bed_capacity':icu_bed_capacity, 'all_bed_usage':all_bed_usage, 'icu_bed_usage':icu_bed_usage})
    return outDF
# mn = GetEquipData(MNFIPS)
# mn
```

```python
# As of now, I just have this function returning its own data frame.

# Returns data from imhe_hospitalizations.csv as a data frame
def GetHospitalizationData(fipsNum, fileName = 'our_data/imhe_hospitalizations.csv'):
    imhe_local_data = pd.read_csv(fileName)
    
    state = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['state']
    
    # Will the StringToList function be called here or later on...?
    allbed_mean = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['allbed_mean']
    allbed_lower = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['allbed_lower']
    allbed_upper = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['allbed_upper']
    
    ICUbed_mean = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['ICUbed_mean']
    ICUbed_lower = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['ICUbed_lower']
    ICUbed_upper = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['ICUbed_upper']
    
    InvVen_mean = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['InvVen_mean']
    InvVen_lower = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['InvVen_lower']
    InvVen_upper = imhe_local_data[imhe_local_data['FIPS'] == fipsNum]['InvVen_upper']
    
    outDF = pd.DataFrame({'FIPS':fipsNum, 'state':state,
                    'allbed_mean':allbed_mean, 'allbed_lower':allbed_lower,
                    'allbed_upper':allbed_upper, 'ICUbed_mean':ICUbed_mean,
                    'ICUbed_lower':ICUbed_lower, 'ICUbed_upper':ICUbed_upper,
                    'InvVen_mean':InvVen_mean, 'InvVen_lower':InvVen_lower,
                    'InvVen_upper':InvVen_upper})
    return outDF
    
#GetHospitalizationData(MNFIPS)
```
